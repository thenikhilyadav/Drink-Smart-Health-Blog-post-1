```python
!pip install matplotlib pillow

import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# ----------------------------
# Data for the bar chart
# ----------------------------
categories = [
    "Crypto\nSpeculation",
    "Patient\nCare",
    "Digital\nHealth Ops",
    "Clinical\nExpansion"
]

before = [90, 50, 55, 30]   # illustrative values
after =  [20, 90, 75, 80]

x = np.arange(len(categories))
width = 0.35

# ----------------------------
# Create figure
# ----------------------------
fig, ax = plt.subplots(figsize=(14, 8))

# Background color
fig.patch.set_facecolor("#f5f7fa")
ax.set_facecolor("#f5f7fa")

# Bars
bars1 = ax.bar(x - width/2, before, width, label="Before Pivot", color="#94a3b8")
bars2 = ax.bar(x + width/2, after, width, label="After Pivot", color="#2563eb")

# ----------------------------
# Titles & labels
# ----------------------------
ax.set_title(
    "KindlyMD Strategic Reset: Patients Over Crypto",
    fontsize=26,
    fontweight="bold",
    pad=20
)

ax.set_ylabel("Strategic Focus Level", fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=12)
ax.set_ylim(0, 100)

# Legend
ax.legend(fontsize=13)

# ----------------------------
# Annotate bars with values
# ----------------------------
def label_bars(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(
            f'{int(height)}%',
            xy=(bar.get_x() + bar.get_width() / 2, height),
            xytext=(0, 5),
            textcoords="offset points",
            ha='center',
            fontsize=11
        )

label_bars(bars1)
label_bars(bars2)

# ----------------------------
# Add key message text
# ----------------------------
plt.figtext(
    0.5, 0.02,
    "Key Insight: $5B crypto strategy reversed to prioritize patient care, stability, and trust",
    ha="center",
    fontsize=14,
    color="#1e293b"
)

# ----------------------------
# Save image
# ----------------------------
file_name = "kindlymd_patient_first_bar_graph.png"
plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.savefig(file_name, dpi=200, bbox_inches="tight")
plt.show()

print(f"âœ… Image created successfully: {file_name}")
print("ðŸ“¥ Download it from the Colab Files panel (left sidebar).")
```
